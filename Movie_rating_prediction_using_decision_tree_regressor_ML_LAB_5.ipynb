```python
#movie rating prediction using decision tree regressor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/ML Dataset/Task 5/tmdb_5000_movies.csv')
dataset.describe()
```

```python
X = dataset[['id', 'vote_average','vote_count']]
Y = dataset[['popularity']]

X_train, X_test , Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
model = DecisionTreeRegressor(random_state = 40)
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)

meansq = mean_squared_error(Y_test, Y_pred)
print("Mean square error", meansq)
```

```python

plt.scatter(Y_test, Y_test, color='Red', label='Actual Ratings')
plt.title('MOVIE PRIDECTED RATING')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.show()
```

```python
import csv
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/ML Dataset/Task 3/diabetes.csv')
dataset.drop_duplicates(inplace=True)
x = dataset['BloodPressure'].values.reshape(-1, 1)
y = dataset['BloodPressure'].values.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 0)
print(x)

```

```python
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(x_train,y_train)

LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='warn', n_jobs=None, penalty='l2',
                   random_state=0, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False)

model = LogisticRegression(solver='liblinear', random_state=0).fit(x,y)
```

```python
model.classes_
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict_proba(x)
```

```python
model.predict(x)
```

```python
confusion_matrix(y, model.predict(x))
```

```python
cm = confusion_matrix(y, model.predict(x))

fig, ax = plt.subplots(figsize=(10, 40))
ax.imshow(cm)
ax.grid(False)
ax.xaxis.set(ticks=(0, 1), ticklabels=('Predicted 0s', 'Predicted 1s'))
ax.yaxis.set(ticks=(0, 1), ticklabels=('Actual 0s', 'Actual 1s'))
ax.set_ylim(1.5, -0.5)
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center', color='red')
plt.show()
```

```python
print(classification_report(y, model.predict(x)))
```

```python
model.score(x, y)
confusion_matrix(y, model.predict(x))

print(classification_report(y, model.predict(x)))
```

```python
#Data set 2 Lung cancer
dataset = pd.read_csv('/lung cancer.csv')
dataset.drop_duplicates(inplace=True)
x = dataset['LUNG_CANCER'].values.reshape(-1, 1)
y = dataset['LUNG_CANCER'].values.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 0)
print(x)



model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(x_train,y_train)

LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='warn', n_jobs=None, penalty='l2',
                   random_state=0, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False)

model = LogisticRegression(solver='liblinear', random_state=0).fit(x,y)
```
